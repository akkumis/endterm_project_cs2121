```python
#Sarsenova Akkumis
#ENDTERM PROJECT
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv(r"C:\Users\akkum\Downloads\Crimes_-_2001_to_Present (1).csv")
```

```python
# Data Preparation (ETL)
# Check for missing values
print("Missing Values:")
print(data.isnull().sum())
```

```python
# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# Data Analysis
# Summary statistics
print("\nSummary Statistics:")
print(data.describe())
```

```python
# Distribution of crimes over the years
plt.figure(figsize=(10, 6))
data['Year'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Crimes Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of crimes by primary type
plt.figure(figsize=(10, 6))
sns.countplot(x='Primary Type', data=data)
plt.title('Distribution of Crimes by Primary Type')
plt.xlabel('Primary Type')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=90)
plt.show()
```

```python
# Trends in crimes by primary type over time
plt.figure(figsize=(12, 8))
sns.lineplot(x='Year', y='ID', hue='Primary Type', data=data)
plt.title('Trends in Crimes by Primary Type Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.legend(title='Primary Type')
plt.show()
```

```python
# Geographic analysis: distribution of crimes by location description
plt.figure(figsize=(12, 8))
data['Location Description'].value_counts().head(20).plot(kind='bar')
plt.title('Distribution of Crimes by Location Description')
plt.xlabel('Location Description')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=90)
plt.show()
```
